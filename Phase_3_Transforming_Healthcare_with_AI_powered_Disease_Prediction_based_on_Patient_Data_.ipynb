```python
# Install libraries
!pip install -q pandas scikit-learn

# Upload file
from google.colab import files
import pandas as pd
import shutil
import streanlit as st

# Upload the file
uploaded = files.upload()

# Rename uploaded file to predefined name
for file_name in uploaded.keys():
    shutil.move(file_name, 'cleaned_disease_dataset.csv')

# Load the dataset
df = pd.read_csv('cleaned_disease_dataset.csv')

print("✅ Dataset loaded successfully!")
df.head()
```

```python
# Split features and target
X = df.drop('prognosis', axis=1)
y = df['prognosis']

# Encode target labels
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Split into train/test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Scale features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Train RandomForest model
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train_scaled, y_train)

# Evaluate
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)
print(f"✅ Model trained with {accuracy*100:.2f}% accuracy!")
```

```python
import pickle

# Save model
with open('model.pkl', 'wb') as f:
    pickle.dump(model, f)

# Save scaler
with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

# Save label encoder
with open('label_encoder.pkl', 'wb') as f:
    pickle.dump(le, f)
```

```python
# Load everything
with open('model.pkl', 'rb') as f:
    model = pickle.load(f)

with open('scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)

with open('label_encoder.pkl', 'rb') as f:
    le = pickle.load(f)

# List of symptoms
symptoms_list = list(X.columns)

# Function to collect user input
def get_user_input():
    print("\nPlease answer Yes or No for the following symptoms:\n")
    user_symptoms = []
    for symptom in symptoms_list:
        ans = input(f"Do you have {symptom.replace('_', ' ')}? (yes/no): ").strip().lower()
        if ans == 'yes':
            user_symptoms.append(1)
        else:
            user_symptoms.append(0)
    return [user_symptoms]

# Function to predict disease
def predict_disease(user_input):
    user_scaled = scaler.transform(user_input)
    pred_encoded = model.predict(user_scaled)
    disease = le.inverse_transform(pred_encoded)
    return disease[0]

# Predict
user_input = get_user_input()
predicted_disease = predict_disease(user_input)

print(f"\n🔴 Based on your symptoms, you might have: {predicted_disease}")
```
